# 1. Missing Values- To find out features that have missing values i.e Nan

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
features_with_na
```

```python
for feature in features_with_na:
    print(feature, "-missing values- ", np.round(a= dataset[feature].isnull().mean(), decimals=4))
```

### we need to find the relationship between missing values and Sales Price

```python
for feature in features_with_na:
    data = dataset.copy()
    #if the value is 1, there exists a null value
    data[feature] = np.where(data[feature].isnull(), 1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.show()
```

###### From the above graphs,the realtionship between the Sales and the null values is clearly visible.
###### Hence, we need to replace these null values with something meaningfull.

```python
print("Id Of Houses: ", len(dataset["Id"]))
```

# 2. Numerical variables

```python
# The feature is not Object"O" is a  numerical feature
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes !="O"]
numerical_features
```

```python
len(numerical_features)
```

```python
dataset[numerical_features].head()
```

```python
year_feature = [feature for feature in numerical_features if "Yr" in feature or "Year" in feature]
year_feature
```

```python
#print the unique values in each year feature

for feature in year_feature:
    print(feature, dataset[feature].unique() )
```

#### Find the realtionship between Year Sold and Sales Price

```python
dataset.groupby("YrSold")["SalePrice"].median().plot()
plt.ylabel("Median House Price")
plt.title("House Price Vs Yr Sold")
```

```python
#Compare the difference between all year feature with Sales Price
for feature in year_feature:
    if feature != "YrSold":
        data = dataset.copy()
        data[feature] = data["YrSold"] - data[feature]
    
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.show()
```

# 3.Distribution of the Numerical Variables

#### Numerical variables are usually of 2 type
#### 1. Continous variable 2. Discrete Variables

```python
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
discrete_feature
```

```python
len(discrete_feature)
```

```python
dataset[discrete_feature].head()
```

## Realtionship between Discrete variables and SalesPrice

```python
for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.ylabel("SalePrice")
    plt.show()
```

## Continous Variables

```python
continous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+["Id"]]
continous_feature
```

```python
len(continous_feature)
```

##### We need to find out the distribution of Continous Values with histograms

```python
for feature in continous_feature:
    data = dataset.copy()
    data[feature].hist(bins= 25)
    plt.title(feature)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
   
```

#### Using logarithamic transformation to convert a skewed data into a normal distributed data

```python
len(data[feature].unique())
```

```python

for feature in continous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data["SalePrice"] = np.log(data["SalePrice"])
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.title(feature)
        plt.ylabel("SalePrice")
        plt.show()
```
